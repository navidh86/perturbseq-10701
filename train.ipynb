```python
train_dataloader = get_dataloader(type="train", majority_fraction=0.01, batch_size=4)
```

```python
# get count of each class in the training data
class_counts = {}
for _, labels in train_dataloader:
    for label in labels:
        label_item = label.item()
        if label_item in class_counts:
            class_counts[label_item] += 1
        else:
            class_counts[label_item] = 1
```

```python
test_dataloader = get_dataloader(type="test", majority_fraction=0.01, batch_size=4)
```

```python
class_counts_test = {}
for _, labels in test_dataloader:
    for label in labels:
        label_item = label.item()
        if label_item in class_counts_test:
            class_counts_test[label_item] += 1
        else:
            class_counts_test[label_item] = 1
```
