## See contents of gff file (annotations on the genome, from which we can get particular gene sequences)

```python
pd.set_option('display.max_columns', None)   # show all columns
pd.set_option('display.max_colwidth', None)  # do NOT truncate long text fields
```

```python
gff = pd.read_csv(
    "GRCh38_latest_genomic.gff",
    sep="\t",
    comment="#",
    header=None,
    names=["seqid", "source", "type", "start", "end", "score", "strand", "phase", "attributes"]
)

# Filter to only genes
genes = gff[gff["type"] == "gene"]
print(genes.head())
```

```python
# Find rows whose attributes mention a particular gene, e.g., ABL1
aatf_rows = genes[genes["attributes"].str.contains("Name=ABL1", case=False, na=False)]
print(aatf_rows)
```

```python
genes.columns
```

### Using seqid, start, end to extract sequence

```python
# !pip install pyfaidx
# !pip install Bio
```

```python
from pyfaidx import Fasta
from Bio.Seq import Seq

genome = Fasta("GRCh38_latest_genomic.fna")
```

```python
seqid = aatf_rows.iloc[0]["seqid"]
start = aatf_rows.iloc[0]["start"]
end = aatf_rows.iloc[0]["end"]
strand = aatf_rows.iloc[0]["strand"]

seq = str(genome[seqid][start-1:end])  # -1 because pyfaidx is 0-based

if strand == "-":
	seq = str(Seq(str(seq)).reverse_complement())
	

print(len(seq))
print(seq[:200])
```

#### Alternative: consider only the promoter region, 2000 bp up and downstream of transcription start site (TSS)

```python
tss = start if strand == "+" else end

promoter_start = max(0, tss - 2000 - 1)  # -1 for 0-based
promoter_end = tss + 2000

promoter_seq = str(genome[seqid][promoter_start:promoter_end])
if strand == "-":
	promoter_seq = str(Seq(str(promoter_seq)).reverse_complement())

print(promoter_seq[:100])
```
