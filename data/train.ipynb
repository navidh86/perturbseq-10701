```python
# check that the three datasets are disjoint
train_set = set(train_dataloader.dataset.df[["tf_name", "gene_name"]].apply(tuple, axis=1))
```

```python
train_set
```

```python
val_set = set(val_dataloader.dataset.df[["tf_name", "gene_name"]].apply(tuple, axis=1))
```

```python
test_set = set(test_dataloader.dataset.df[["tf_name", "gene_name"]].apply(tuple, axis=1))
```

```python
assert (train_set - val_set == train_set) and (train_set - test_set == train_set) and (val_set - test_set == val_set)
```
